```python
import pandas as pd
```

```python
df=pd.read_csv('../dataset/orders.csv')
df
```

```python
df.rename(columns={
    'Order Id': 'order_id',
    'Order Date': 'order_date',
    'Ship Mode': 'ship_mode',
    'Segment': 'segment',
    'Country': 'country',
    'City': 'city',
    'State': 'state',
    'Postal Code': 'postal_code',
    'Region': 'region',
    'Category': 'category',
    'Sub Category': 'sub_category',
    'Product Id': 'product_id',
    'cost price': 'cost_price',
    'List Price': 'list_price',
    'Quantity': 'quantity',
    'Discount Percent': 'discount_percent'
}, inplace=True)
```

```python
df = df.set_index("order_id")
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=["ship_mode"])
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df['sub_category'].unique()
```

```python
df['ship_mode'].unique()
```

```python
df['final_customer_price']=(df['list_price']*df['quantity'])*(1-(df['discount_percent']/100))
```

```python
df['profit']=df['final_customer_price']-(df['cost_price']*df['quantity'])
```

```python
df
```

```python
profit_category=df.groupby('sub_category')['profit'].sum().reset_index(name="Total Profit")
```

```python
profit_category
```

```python
profit_category.to_csv('../dataset/prepared/profit_category.csv')
```

```python
df['profit'].sum()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce', dayfirst=True)
monthly_profit=df.groupby(df['order_date'].dt.to_period('M'))['profit'].sum().reset_index(name='monthly_price')
monthly_profit
```

```python
monthly_profit.to_csv('../dataset/prepared/monthly_profit.csv')
```
